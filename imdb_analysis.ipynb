### 1.5. Create a movie_duration_category column, which will contain three categories depending on the film length:
• Category "1. <90" if the film is less than 90 minutes long

• Category "2. 90–120" if the film is between 90 minutes and two hours long (inclusively)

• Category "3. >120" if the film is more than two hours long


To do so, we are going to use the Numpy library function `np.where`, which allow us to use multiple if statements in a pd.Series. Since it is a 3 step process, we need to use the np.where twice - one time to identify movies with less than 90 minutes, and once again to distinguish between movies with more than 120 minutes and the movies between 90 and 120 minutes.

We could also use list comprehension or create a function and .apply it to the column. Doing so we miss the power of vectorization that Numpy offers it's users

```python
df_imdb['movie_duration_category'] = np.where(
        df_imdb['duration'] < 90, '1. <90',
        np.where(df_imdb['duration'] > 120, '3. >120','2. >90-120')
)
```

### 1.8. The plot_keywords column holds keywords characterizing the film's plot. Using the data in this column, create a column called movie_plot_category, to contain four categories depending on the key words in the column:

• Category "love_and_death" if the keywords include both "love" and "death"

• Category "love" if the keywords include the word "love"

• Category "death" if the keywords include the word "death"

• Category "other" if the keywords do not meet the conditions above

To do this categorization, we will use the `pd.str.contains()` function to map keywords and sort the movies accordingly. To do so, we will return to use the **df_imdb** Dataframe. 

Once again, we will use `np.where`, but a simpler solution with function will also be displayed.

```python
df_imdb['plot_keywords']
```

```python
df_imdb['movie_plot_category'] = np.where(
    #love_and_death
    df['plot_keywords'].str.contains("love") & df['plot_keywords'].str.contains("death"), "love_and_death",
    #love
    np.where(df['plot_keywords'].str.contains("love"), "love",
    #death
             np.where(df['plot_keywords'].str.contains("death"), "death", 
                      "other"))
             
             )
```

```python
## using a function and .apply():

def categorize_keywords(keywords):
    if keywords.str.contains("love") and keywords.str.contains("death"):
        return "love_and_death"
    elif keywords.str.contains("love"):
        return "love"
    elif keywords.str.contains("death"):
        return "death"
    else:
        return "other"
#Not running, but then you use .apply() to 'literaly' apply the function to the column.
#df_imdb['movie_plot_category_2'] = df_imdb['plot_keywords'].apply(categorize_keywords)
```

### 1.9. The imdb_score column shows a viewer rating for the film. Build a table to reflect the average rating of films depending on which movie_plot_category category they belong to.

Using another method to create an grouping table between two columns is to use .groupby() and the desired calculation afterwards. We're also rounding up to two decimals.

```python
df_avg_rating = pd.DataFrame(df_imdb.groupby('movie_plot_category')['imdb_score'].mean().round(2))
df_avg_rating
```
